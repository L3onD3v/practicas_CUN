# 2. Limpieza Básica (40%)

```python
#Identifica la columna STATE. Llena los valores vacíos (nulos)
#con el texto "Desconocido".
# Vamos a ver todos los nombres de las columnas
print(df.columns.tolist())
```

```python
# Rellenar los espacios vacíos de la columna STATE con "Desconocido"
df['STATE'] = df['STATE'].fillna('Desconocido')

# Ver el resultado (primeras 10 filas)
print(df['STATE'].head(10))
```

```python
#Vamos a verificar si hay columnas duplicadas
duplicados = df.duplicated().sum()
print(f"Hay {duplicados} filas duplicadas")
```

```python
#Ahora vamos a sacar el promedio de la columnas SALES
promedio_sales = df['SALES'].mean()
print(f"El promedio de la columna SALES es:{promedio_sales:.2f}")
```

```python
# Mostrar las filas que están duplicadas
filas_duplicadas = df[df.duplicated(keep=False)]
print(filas_duplicadas)
```

```python
# Eliminar duplicados y guardar el resultado
df_sin_duplicados = df.drop_duplicates()

# Ver cuántas filas quedaron
print(f"Filas originales: {len(df)}")
print(f"Filas después de eliminar duplicados: {len(df_sin_duplicados)}")
print(f"Se eliminaron {len(df) - len(df_sin_duplicados)} filas duplicadas")
```

```python
# Agrupar por PRODUCTLINE y sumar las ventas (SALES)
ventas_por_producto = df.groupby('PRODUCTLINE')['SALES'].sum()

# 5. Mostrar los resultados en números
print("=== VENTAS TOTALES POR LÍNEA DE PRODUCTO ===")
print(ventas_por_producto)
print("\n")
```

# 3. Visualización de Resultados (30%)

```python
ventas_por_producto = df.groupby('PRODUCTLINE')['SALES'].sum()
ventas_por_producto.plot(kind='bar')
plt.title('Ventas por Línea de Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.xticks(rotation=90)
plt.show()
```
